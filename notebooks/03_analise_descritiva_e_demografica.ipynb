# Análise de Dengue - Pirâmide Etária

- Este notebook realiza a análise dos casos de dengue por faixa etária e sexo,
construindo pirâmides etárias para visualização dos dados.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_deng_norm = pd.read_csv('../data/final/df_deng_norm.csv')
```

## 1. Pirâmide Etária Geral

- Agrupamos os casos por faixa etária e sexo para construir a pirâmide populacional.
Os valores do sexo masculino (M) são invertidos para criar o efeito espelhado da pirâmide.

```python
piramide = df_deng_norm.groupby(
    ['faixa_etaria', 'sexo']
).size().unstack(fill_value=0)
```

### Visualização da Pirâmide

```python
piramide['M'] = piramide['M'] * -1
piramide.plot(
    kind='barh',
    figsize=(8,6)
)
plt.title('Pirâmide etária dos casos de dengue')
plt.xlabel('Número de casos')
plt.ylabel('Faixa etária')
plt.show()
```

## 2. Pirâmide Etária por Ano de Notificação

Aqui agrupamos também pelo ano de notificação para comparar a distribuição
dos casos ao longo dos anos.

```python
piramide_ano = df_deng_norm.groupby(
    [ 'ano_notificacao', 'faixa_etaria', 'sexo']
).size().unstack(fill_value=0)
piramide_ano['M'] = piramide_ano['M'] * -1

piramide_ano
```

## 3. Pirâmide Etária - 2022

Filtramos os dados apenas para o ano de 2022 para uma análise mais detalhada.

```python
piramide_2022 = piramide_ano.loc[2022]
piramide_2022.plot(kind='barh',figsize=(8,6))
```

# Análise de Dengue - Distribuição por sexo (percentual)

```python
distribuicao = df_deng_norm.groupby(
    ['sexo']
).size()

distribuicao_df = distribuicao.reset_index(name='total')
distribuicao_df['percentual'] = (distribuicao_df['total'] / distribuicao_df['total'].sum() * 100).round(2)
distribuicao_df
```

```python
distribuicao_df.plot.pie(y='total',labels=distribuicao_df['sexo'], figsize=(5, 5), autopct='%1.1f%%')
```

# Análise de Dengue - Evolução dos casos (desfecho clínico)

```python
evolucao = df_deng_norm.groupby(
    ['evolucao_caso']
).size()

evolucao_df = evolucao.reset_index(name='total')

import json 
with open('../src/dicionario-evolucao.json', 'r') as f:
    dic_evol_dg = json.load(f)
    
evolucao_df['percentual'] = (evolucao_df['total'] / evolucao_df['total'].sum() * 100).round(2)
evolucao_df['evolucao_caso'] = evolucao_df['evolucao_caso'].astype(str).map(dic_evol_dg)
evolucao_df
```

```python
ax = evolucao_df.plot.bar(x='evolucao_caso', y='total', rot=30)

for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 300,
            int(bar.get_height()), ha='center', fontsize=9)


plt.tight_layout()
plt.show()
```

# Análise de Dengue - Evolução por município

```python
codigo_ilheus_itabna = {'291360': 'Ilhéus', '291480': 'Itabuna'}

evol_municipio = df_deng_norm.groupby(
    ['municipio_notificacao', 'evolucao_caso']
).size()


evol_municipio_df = evol_municipio.reset_index(name='total')
evol_municipio_df['evolucao_caso'] = evol_municipio_df['evolucao_caso'].astype(str).map(dic_evol_dg)
evol_municipio_df['percentual'] = (evol_municipio_df['total'] / evol_municipio_df.groupby('municipio_notificacao')['total'].transform('sum') * 100).round(2)
evol_municipio_df['municipio_notificacao'] = evol_municipio_df['municipio_notificacao'].astype(str).map(codigo_ilheus_itabna)
evol_municipio_df
```
