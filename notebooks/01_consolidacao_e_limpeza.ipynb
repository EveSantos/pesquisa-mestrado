## Padronizar datas

- Primeiro criar uma função
- Transforma o valor recebido para string
- Caso tenha 8 caracteres e nao esteja no formato YYYY-MM-DD vai formatar corretamente


```python
def parse_data(val):
    val = str(val).strip()
    if len(val) == 8 and '-' not in val:  # formato YYYYMMDD
        return pd.to_datetime(val, format='%Y%m%d')
    return pd.to_datetime(val)  # formato já correto

df_deng_total_ren['data_notificacao'] = df_deng_total_ren['data_notificacao'].apply(parse_data)
```

### Definir faixa etária para os dados analisados para facilitar leitura e interpretação

```python
bins = [0,1,5,10,20,40,60,80,120]
labels = ['<1','1-4','5-9','10-19','20-39','40-59','60-79','80+']

df_deng_total_ren['faixa_etaria'] = pd.cut(df_deng_total_ren['idade_anos'],bins=bins, labels=labels)
df_deng_total_ren
```
