## Inicio do projeto

Como os arquivos ja foram convertidos em csv, vou fazer a leitura de cada csv e separar os dados de Ilhéus e Itabuna

```python
import pandas as pd
```

### Ler e salvar dados de 2014 a 2019

```python
# Ler dados de 2014 a 2019
deng_dados14 = pd.read_csv('./data/DENGBR14.csv')
deng_dados15 = pd.read_csv('./data/DENGBR15.csv')
deng_dados16 = pd.read_csv('./data/DENGBR16.csv')
deng_dados17 = pd.read_csv('./data/DENGBR17.csv')
deng_dados18 = pd.read_csv('./data/DENGBR18.csv')
deng_dados19 = pd.read_csv('./data/DENGBR19.csv')
```

```python
# Ler apenas os dados de ilhéus e itabuna do ano de 2014 a 19
codigo_ilheus_itabna = [291360,291480]

deng_dados14_ba = deng_dados14[deng_dados14['ID_MUNICIP'].isin(codigo_ilheus_itabna)]
deng_dados15_ba = deng_dados15[deng_dados15['ID_MUNICIP'].isin(codigo_ilheus_itabna)]
deng_dados16_ba = deng_dados16[deng_dados16['ID_MUNICIP'].isin(codigo_ilheus_itabna)]
deng_dados17_ba = deng_dados17[deng_dados17['ID_MUNICIP'].isin(codigo_ilheus_itabna)]
deng_dados18_ba = deng_dados18[deng_dados18['ID_MUNICIP'].isin(codigo_ilheus_itabna)]
deng_dados19_ba = deng_dados19[deng_dados19['ID_MUNICIP'].isin(codigo_ilheus_itabna)]
```

```python
# Salvar novos csvs de 14 a 19
deng_dados14_ba.to_csv('data-filtred/deng_dados14_ba.csv', index=False)
deng_dados15_ba.to_csv('data-filtred/deng_dados15_ba.csv', index=False)
deng_dados16_ba.to_csv('data-filtred/deng_dados16_ba.csv', index=False)
deng_dados17_ba.to_csv('data-filtred/deng_dados17_ba.csv', index=False)
deng_dados18_ba.to_csv('data-filtred/deng_dados18_ba.csv', index=False)
deng_dados19_ba.to_csv('data-filtred/deng_dados19_ba.csv', index=False)
```

### Ler dados de 2023

Como os dados de 2024 são muito grandes, foi necessario fazer a leitura por chunks, facilitando e consumindo menos memoria

```python
chunks = pd.read_csv('./data/DENGBR24.csv', chunksize=100000, low_memory=False)
deng_dados24 = pd.concat(chunks)
```

```python
deng_dados24_ba = deng_dados24[deng_dados24['ID_MUNICIP'].isin(codigo_ilheus_itabna)]
```
